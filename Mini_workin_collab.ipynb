```
import os
import zipfile # Chemin du dossier à télécharger
folder_path = "/content/Project_NLP_MVA_23/Outputs_collab"

# Nom de l'archive zip
zip_name = "mon_archive.zip"

# Création de l'archive zip
with zipfile.ZipFile(zip_name, "w", zipfile.ZIP_DEFLATED) as zip:
  for root, dirs, files in os.walk(folder_path):
    for file in files:
      zip.write(os.path.join(root, file))
```
