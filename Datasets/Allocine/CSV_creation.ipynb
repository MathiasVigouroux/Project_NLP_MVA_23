```python
import pyarrow.parquet as pq
import json
import pandas as pd 

# Read the parquet file
table_train = pq.read_table('train.parquet')
# Convert to a dictionary
data_train = table_train.to_pydict()

data_train = pd.DataFrame(data_train)

display(data_train.head())
data_train.shape 
```

```python
train = data_train.sample(32,random_state=2280) # as in the paper, the train data is made out of 32 examples from the train data
train['idx'] = train.index #they also use an idx column in their train data
train.to_csv("train.csv", index=False)
display(train)
train.shape
```

```python
train_test = pd.read_csv("train.csv") # ok it is reading it correctly.
train_test
```

```python
unlabeled=data_train.drop(train.index)
unlabeled['idx'] =unlabeled.index
unlabeled = unlabeled.sample(10000,random_state=2280) # the unlabled data would be of 159968 which is too large by a factor 10
unlabeled.drop(['label'], axis=1, inplace=True) # fron the train data set the unlabeled dataset is created by removing the label
unlabeled.to_csv('unlabeled.csv', index=False)
display(unlabeled.head())
unlabeled.shape
```
