```python
import json 
import pandas as pd


path = "/Users/mathias/Desktop/Project_NLP_MVA_23/Datasets/HuCoPA/data/train_initial.json"



with open(path, encoding='utf8') as json_file:
    json_data = json.load(json_file)


traininit_data = pd.DataFrame(json_data)
traininit_data.index = traininit_data['id']

#traininit_data.drop('id',axis=1,inplace=True)


traininit_data
```

```python
train_data = traininit_data.sample(32)
train_data.head()
```

```python
json_list = []

# Loop through each row in the DataFrame and create a JSON object for each row
for idx, row in train_data.iterrows():
    json_obj = {"id": row["id"], "question": row["question"], "label":row["label"], "premise":row["premise"], "choice1":row["choice1"], "choice2":row["choice2"]}
    json_list.append(json_obj)

print(json_list)
filename = "/Users/mathias/Desktop/Project_NLP_MVA_23/Datasets/HuCoPA/data/train.json"

# Open the file and write the JSON string to it
with open(filename, 'w', encoding='utf8') as json_file:
    json.dump(json_list, json_file, ensure_ascii=False)
```

```python
mask = ~traininit_data.index.isin(train_data.index)

unlabelled_data = traininit_data[mask]
unlabelled_data.drop('label',axis=1, inplace=True)
unlabelled_data
```

```python
json_list = []

# Loop through each row in the DataFrame and create a JSON object for each row
for idx, row in unlabelled_data.iterrows():
    json_obj = {"id": row["id"], "question": row["question"], "premise":row["premise"], "choice1":row["choice1"], "choice2":row["choice2"]}
    json_list.append(json_obj)

print(json_list)
filename = "/Users/mathias/Desktop/Project_NLP_MVA_23/Datasets/HuCoPA/data/unlabelled.json"

# Open the file and write the JSON string to it
with open(filename, 'w', encoding='utf8') as json_file:
    json.dump(json_list, json_file, ensure_ascii=False)
```
